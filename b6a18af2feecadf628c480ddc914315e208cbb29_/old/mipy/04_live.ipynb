## break a continue
- break prerusi cyklus a pokracuje zbytkem programu
- continue skoci na dalsi iteraci


````python
for i in range(5):
    if i == 2:
        break
    print(i)
    
print()

for i in range(5):
    if i ==2:
        continue
    print(i)
````

## List a dict comprehension

````python
lst = list(range(10))
lst2 = []
for x in lst:
    lst2.append(2*x)
print(lst2)
````

````python
lst2 = [sqrt(x) for x in lst]
print(lst2)
````

````python
lst3 = [x/2 for x in lst if x % 2 == 0]
print(lst3)
````

````python
lst4 = ["Even" if x%2==0 else "odd" for x in lst]
print(lst4)
````

````python
opts = {
    "opt_a" : 1,
    "opt_b" : 2,
    "opt_c" : 3
}

opts2 = {key[-1]:(val-1) for key, val in opts.items()}
print(opts2)
````

## Typy argumentu funkci
- pozicni
- arbitrary (variable length args)
- keyword arguments (kwargs)

````python
def funkce(arg1 = 11, arg2 = 77):
    print(arg1, arg2)
    
funkce(1, 2)
funkce(2, 1)
funkce()
funkce(arg2 = -1)
funkce(1)
````

````python
def funkce(*arg):
    print(type(arg), arg)
    # return [2*x for x in arg]
    
print(funkce(1, 2, 3, True))
````

````python
def funkce(**kwargs):
    print(kwargs)
    
funkce(a = 2, b = True)
````

````python
def slozita_funkce(x, y, *arg, **kwargs):
    print("pos. args: ", x, y)
    print("arg tuple: ", arg)
    print("kwargs: ", kwargs)
    
slozita_funkce(2, 1, "tuple1", "tuple2", kwarg1 = True, kwarg2 = "sadlo")
````
