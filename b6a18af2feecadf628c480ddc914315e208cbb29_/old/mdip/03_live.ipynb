# list

```python
lst = [1, 2.0, True, "neco"]

print(lst[::-1])

lst.append(2)
print(lst)
x = lst.pop(2)
print(lst, x)
```

```python
# list je mutable

x = [1, 2]
y = x
print(x is y)

x[0] = 0
print(x is y)
print(x, y)
```

```python
# remove outliers
def remove_outliers(x):
    for i in range(len(x)):
        if x[i] > 3:
            x[i] = 0
    return x

x = [1, 2, 5, 3]
y = remove_outliers(x)

print(x, y)
```

```python
# kopie

from copy import copy

x = [1, 2]
y = copy(x)
print(x is y)

x[0] = 0
print(x is y)
print(x, y)
```

```python
x = [1, 2]
y = [*x]
z = x[:]
print(x is y)
print(*x) # print(x[0], x[1])
print(x is z)
```

```python
from copy import copy, deepcopy
x = [
    [1, 2],
    [3, 4]
]
# y = copy(x)
y = deepcopy(x)

x[0][0] = 0
print(y)
print(x is y)
print(x[0] is y[0])
```

## iterace (iterate over a list/tuple...)

```python
lst = [1, 2,3, 4,5, 6,7]
lst2 = ["a", "b", "c"]

for x in lst:
    print(x)
    
for i, x in enumerate(lst):
    print(i, x)
    
for a, b in zip(lst, lst2):
    print(a, b)
    
for x in zip(lst, lst2):
    print(x, type(x))
```

## Dictionary (slovnik, hash)

```python
user = {
    "name" : "Erwin Schrodinger",
    "phone" : 123456789
}

print(user)
print(user["name"])
key = "name"
print(user[key])

for x in user:
    print(x, user[x])
    
for key, val in user.items():
    print(key, val)
```

```python
user = {
    "name" : "Erwin Schrodinger",
    "phone" : {
                "cell" : 123456789,
                "home" : 987654321
                }
}

outstr = """
Name: {name}
Telephone:
    cell: {phone[cell]}
    home: {phone[home]}
""".format(**user)
print(outstr)
print(f"{user['name']}")
```

# Prvocisla

Ověřte, že dané číslo je prvočíslo.

```python
x = 1

is_prime = True
if x == 1:
    is_prime = False
else:
    for i in range(2, x):
        if x % i == 0:
            is_prime = False
        
if is_prime:
    print(f"{x} is a prime")
else:
    print(f"{x} is not a prime")
```
