## Mutable vs. Immutable

- vše je objekt
- objekty seskupují "data" a funkce na nich ("metody")
- každý objekt má `id`, `type`, `value`

````python
formatstr = "type({}) = {}, id(a) = {}, a = {}"
def myprint(varname):
    var = eval(varname)
    print(formatstr.format(varname, type(var), id(var), var))

a = 911
b = -13.6


myprint("a")
a += 10000
myprint("a")
myprint("b")

c = [a, b]
myprint("c")
c.append(True)
myprint("c")
````

Immutable typy
- `int`
- `float`
- `bool`
- `string`
- `complex`
- `frozen set`
- `tuple`
- `range`

Mutable typy
- `list`
- `set`
- `dict`

### vytváření kopií

````python
# int je immutable
a = 2
b = a

print(a is b)

b += 1
print(a is b)
print(a, b)
````

````python
# list je mutable
c = [2, 2]
d = c

print(c is d)

d[1] += 1

print(c is d)
print(c, d)
````
