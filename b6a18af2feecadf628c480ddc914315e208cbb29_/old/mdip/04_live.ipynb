break - ukonci chod cyklu a pokracuje zbytkem
continue - prerusi aktualni iteraci a prejde k dalsi

````python
i = 0
while i < 5:
    if i == 2:
        i += 1
        continue
    print(i)
    i += 1
````

## Switch
V mnoha jazycích existuje konstrukce `switch`, která nahrazuje sérii `if-elif-...-elif-else`. Příklad v C:

```c

if (number == 1) {
    do_thing_1();
}
else if (number == 2) {
    do_thing_2();
}
//...
else if (number == k) {
    do_thing_k();
}
else {
    do_default_thing();
}


switch (number) {
    case 1:
        do_thing_1();
        break;
    case 2:
        do_thing_2();
        break;
    \\ ...
    case k:
        do_thing_k();
        break;
    default:
        do_default_thing();
}
```

V Pythonu nic takového není. Ale můžeme předstírat, že je.

````python
from math import sqrt, exp, log, log2

number = 7

switcher = {
    1: sqrt,
    2: exp, 
    3: log
}

func = switcher.get(number, log2)
func(2)
````

## List a dict comprehension

````python
lst = list(range(10))
lst2 = []
for x in lst:
    lst2.append(2*x)
print(lst2)
# not pythonic
````

````python
lst2 = [x * 2 for x in lst]
lst3 = [x for x in lst if (x % 2 != 0)]
lst4 = ["Even" if (x % 2 == 0) else "Odd" for x in lst]
print(lst4)
````

````python
# dict comprehension
opts = {
    "opt_a" : 1,
    "opt_b" : 2, 
    "opt_c" : 3
}
opts2 = {key[-1]:val for key, val in opts.items()}
print(opts2)
dct = {i:i**2 for i in range(10)}
print(dct)
````

## Argumenty funkci
- positional arguments
- arbitrary length (variable length args)
- keywod arguments (kwargs)

````python
def funkce(a, b, c):
    print(a, b, c)
    
funkce(2, 1, 2)
funkce(b = 1, a = 2, c= 1)
funkce(1, c = 2, b=3)
````

````python
def funkce(*args):
    for x in args:
        print(x)
    
funkce(1, 2, 3, True, "a")
````

````python
def funkce(**kwargs):
    print(type(kwargs), kwargs)
    
funkce(a = 1, b = True)
````

````python
def slozita_funkce(x, y, *args, **kwargs):
    print("pos. args: ", x, y)
    print("var. args: ", args)
    print("kwargs: ", kwargs)
    
kwargs = {
    "kwarg1" : "prvni",
    "kwarg2" : 2.0
}
args = (True, "a")
slozita_funkce(1, 2, True, "a", kwarg1 = "prvni", kwarg2 = 2.0)
slozita_funkce(1, 2, *args, **kwargs)
````
