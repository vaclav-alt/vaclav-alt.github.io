## ggplot

```
data_k_vykresleni + prirazeni_os + geometricke_objekty + dalsi_objekty
```

V Pythonu ggplot implementovan v balku `plotnine`
```
pip3 install plotnine
```

````python
import pandas as pd
from plotnine import *

df = pd.read_csv("nakazeni-vyleceni-umrti-testy.csv")
# df = df[::20]
df.set_index(['datum'], drop = True, inplace = True)
df.index = pd.to_datetime(df.index)
new_names = ["inf", "cur", "dead", "tests", "ag", "d_inf", "d_cur", "d_dead", "d_tests", "d_ag"]
df.columns = new_names
````

````python
df2 = df.resample("10D").agg({
    "dead" : "sum",
    "d_dead" : "mean"
})[:-1]
p = (
    ggplot(df2)
    + aes(x = df2.index, y = "dead")
    + geom_line(color = "red")
    + xlab("Nas vlastni popisek osy x")
    + labs(y = "Kumulativni pocet umrti", x = "daum", title = "graf")
    + theme(axis_text_x = element_text(angle = 45, hjust = 1))
    #+ theme_bw()
)
themes = [theme_xkcd, theme_bw]

for the in themes:
    print(p + the())
````

## Scatterplot

````python
import matplotlib.pyplot as plt
import numpy as np

data = {
    'a' : np.arange(50),
    'c' : np.random.randint(0, 50, 50),
    'd' : np.random.randn(50) # 50 nahodnych cisel mezi s normalnim (Gaussovym) rozdelenim
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 5

p = (
    ggplot()
    + aes(x = data['a'], y = data['b'])
    + geom_point(size = data['d'], mapping = aes(color = data['c']))
)
p + labs(color = "Barva") + theme_bw()
````
