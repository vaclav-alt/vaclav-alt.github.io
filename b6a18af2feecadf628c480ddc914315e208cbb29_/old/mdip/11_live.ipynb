## XML - eXtensible Markup Language

```python
with open("data/sample.xml", "r") as f:
    print(f.read())
```

```python
import xml.etree.ElementTree as ET

tree = ET.parse("data/sample.xml")
root = tree.getroot()

for child in root:
    print(child.attrib["name"])
    for grandchild in child:
        print(grandchild.text)
```

## .ini

```python
import configparser as cp
config = cp.ConfigParser()
config.read("data/sample.ini")

for section in config:
    print(section)
    for entry in config[section]:
        print(entry)
    print()
    
print(config["DEFAULT"]["var"])
```

## Webscraping a beautiful soup
- `pip3 install bs4 lxml html5lib`

```python
# stazeni webove stranky - pomoci baliku requests (pip3 install requests)

import requests

r = requests.get("http://vaclav-alt.github.io/data/index.html")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text)
print(soup.title)
div = soup.find(id="content")
print(div)

for table in soup.find_all("table"):
    print(table.caption.text)

tables = soup.find_all("table")
tbl = tables[0]
rows = tbl.find_all("tr")
print(rows)
```

### HTTP, API
- API - application programming interface
- HTTP - hyper text transfer protocol
- REST API - - REST API -

```python
import requests # zpostredkovava komunikaci pred HTTP
import json
import time


url = "https://sv443.net/jokeapi/v2/joke/Any"
url = "https://v2.jokeapi.dev/joke/Any?blacklistFlags=nsfw,sexist,explicit&type=twopart"

r = requests.get(url)
if r.status_code == 200: # vse v poradku
    data = json.loads(r.text)
    
    if (data["type"] == "single"):
        print(data["joke"])
    else:
        print(data["setup"])
        time.sleep(2)
        print(data["delivery"])
```
