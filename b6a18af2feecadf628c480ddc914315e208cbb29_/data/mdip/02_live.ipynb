## Mutability / immutability

```python
a = 400
print(id(a))
a += 1 # a = a + 1
print(id(a))
```

```python
lst1 = [1, 2, 3]
lst2 = lst1

lst1 is lst2
lst1.append(4)
print(lst1, lst2, lst1 is lst2)
```

```python
a = 1.2
b = a
a is b
a += 1
a is b
```

Immutable typy: int, float, bool, string, complex, frozen set, tuple, range
Mutable: list, set, dict

```python
def f(a, b=2):
    return a + b

f(1, 3)
```

```python
def process(x, lst=[]):
    # process x
    lst.append(x)
    return lst

lst = process(1)
lst2 = process(2)
print(lst, lst2)
lst is lst2
```

### Historické okénko, teplota a úskalí floating point aritmetiky

Německý fyzik Gabriel Fahrenheit navrhl teplotní stupnici se dvěma skvěle zvolenými referenčními body. Teplota 0 °F je nejnižší teplota, jakou se podařilo Fahrenheitovi naměřit (v roce 1724) ve směsi chloridu amonného, vody a ledu. Druhým referenčním bodem pak je *normální* teplota lidského těla, fixovaná na 98 °F. Nyní se jako referenční body používají bod mrazu vody (32 °F) a bod varu vody (212 °F), čemuž odpovídají převodní vztahy

$$
 T_C = (T_F - 32) \frac{5}{9}
$$

$$
 T_F = \frac{9}{5} T_C + 32
$$

```python
def get_fahrenheit(tc):
    return tc * 9. / 5. + 32.

def get_celsius(tf):
    return (tf - 32.) * 5. / 9.

def get_celsius2(tf):
    return tf * 5. / 9. - 32. * 5. / 9.

tf = 32.6
tc1 = get_celsius(tf)
tc2 = get_celsius2(tf)
print(tc1)
print(tc2)
tc1 == tc2
```

```python
# floaty vzdy porovnavame takto:

abs(tc1 - tc2) <= 1e-13 # 1e-13 = 1 * 10^(-13)
```

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if (n % i) == 0:
            return False
    return True
    
def get_primes(numbers):
    primes = []
    for num in numbers:
        if is_prime(num):
            primes.append(num)
    return primes

is_prime(4)
get_primes([1, 2, 3, 4, 5])
```

```python
# filter(bool_fce, container) -> container

def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if (n % i) == 0:
            return False
    return True
    
def get_primes(numbers):
    return list(filter(is_prime, numbers)) # very pythonic, functional programming

is_prime(4)
get_primes([1, 2, 3, 4, 5])
```

## Formátovaný výstup



```python
# c-style format

x = 2
y = 4.4
z = "lenochod"

out_str = "x=%d, y=%f, z=%s" % (x, y, z)
print(out_str)
```

```python
# .format

x = 2
y = 4.4
z = "lenochod"

out_str = "x={}, y={}, z={}".format(x, y, z)
print(out_str)
```

```python
# f-string

x = 2
y = 4.4
z = "lenochod"

out_str = f"x={x:+05d}, y={y:+10.7f}, z={z}"
print(out_str)

a = 0.546789
print(f"{a:7.2%}")

```
