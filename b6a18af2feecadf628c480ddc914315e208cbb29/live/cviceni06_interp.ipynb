## navratova hodnota a jeji typ

```python
def return_nothing():
    pass

def return_int():
    return 2

def return_float():
    return 1.0

def return_something(x):
    if x > 0:
        return 1.0
    else:
        return None

# None ~ null
a = return_nothing()
b = return_int()
c = return_float()
d = return_something(-2)
print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
```

## Argumenty / parametr

```python
def funkce(x, y):
    print("x = {}, y = {}".format(x, y))
    
funkce(1, 2)
funkce(2, 1)
funkce(x = 3, y = 4)
funkce(y = 4, x = 3)
funkce(5, y = 6)
funkce(x = 5, 6)
```

```python
# keyword arguments
def funkce3(**kwargs):
    print(type(kwargs))
    for key, val in kwargs.items():
        print(key, val)
    
funkce3(x = 4, y = True, z = "string")
```

```python
def add(x, y, **kwargs):
    val = x + y
    print(kwargs.get("print"))
    if kwargs.get("print", False):
        print(val)
    return val

a = add(1,2)
```

```python
# vsechno dohromady

def slozita_funkce(x, y, *argv, **kwargs):
    print("positional arguments: ", x, y)
    print("arguments tuple:", argv)
    print("keyword arguments: ", kwargs)
    
slozita_funkce(2, 1, "tuple1", "tuple2", "tuple3", True, 2, kwarg1 = True, kwarg2 = "jeleni pastika")
```

## Rekurze

rekurze: viz rekurze

### Faktoriál

V kombinatorice se často užívá symbolu $n!$, čteme $n$-faktoriál. Je definován takto:

$$
n! = \prod\limits_{i=1}^{n} i = n \cdot (n-1) \cdot \dots \cdot 2 \cdot 1 = n \cdot (n-1)!
$$

Např.
$$
2! = 2\\
3! = 6\\
4! = 24
$$

Dodatečně se obvykle definuje $0! = 1$.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def measure_time(n, fun):
    time = %timeit -q -n 100000 -o fun(n)
    return sum(time.timings) / len(time.timings)

# rekurzivni implementace faktorialu
def faktorial1(n):
    if n > 1:
        return n * faktorial1(n-1)
    else:
        return 1

def faktorial2(n):
    if n > 1:
        res = 1
        for i in range(2, n+1):
            res *= i
        return res
    return 1

nrange = range(1, 21)
times1 = [measure_time(n, faktorial1) for n in nrange]
times2 = [measure_time(n, faktorial2) for n in nrange]
times3 = [measure_time(n, math.factorial) for n in nrange]
times4 = [measure_time(n, np.math.factorial) for n in nrange]

# plt.plot(x, y)
plt.plot(nrange, times1, "-o", label="rekurze")
plt.plot(nrange, times2, "-o", label="for loop")
plt.plot(nrange, times3, "-o", label="math.factorial")
plt.plot(nrange, times4, "-o", label="numpy.math.factorial")

# zatim ignorujte
plt.xticks(nrange)
plt.xlabel("n")
plt.ylabel("time(n!) / s")
plt.legend()
plt.yscale('log')
plt.show()
```

```python
k = 1
def do_something():
    l = 1
    def change_l():
        nonlocal l
        l += 1
    def change_k():
        global k
        k += 1
        
    print(k, l)
    change_k()
    print(k, l)
    change_l()
    print(k, l)

do_something()
```
