```python
# list comprehension

from random import sample
# sample(container, N)
ceny_s_dph = sample(range(1000), 10)
dph = 1.21

ceny_bez_dph = []
for x in ceny_s_dph:
    y = round(x / dph, 2)
    ceny_bez_dph.append(y)
print(ceny_bez_dph)
print(ceny_s_dph)
```

```python
# list comprehension
# results = [expression(x) for x in container]
ceny_bez_dph = [x / dph for x in ceny_s_dph]
print(ceny_bez_dph)
```

```python
print("bez dph\t| s dph")
print("=" * 18)
for i in range(len(ceny_bez_dph)):
    print("{0:7.2f}\t|{1:7.2f}".format(ceny_bez_dph[i], ceny_s_dph[i]))
```

```python
# zip(cont1, cont2, cont,....contN) -> (x1, x2, x3,...xN)
print("bez dph\t| s dph")
print("=" * 18)

for base, taxed in zip(ceny_bez_dph, ceny_s_dph):
    print("{0:7.2f}\t|{1:7.2f}".format(base, taxed))
```

```python
import matplotlib.pyplot as plt
from math import sqrt
# chceme nakreslit funkci: sqrt(x) + 1
def op(x):
    return sqrt(x) + 1

def np_op(x):
    return np.sqrt(x) + 1

vec_op = np.vectorize(op)

x = [(0.0 + i * 0.01) for i in range(101)] # definicni obor
%timeit y = [op(i) for i in x]


npx = np.linspace(0., 1.0, 101)
%timeit npy = np_op(npx)

%timeit npy2 = vec_op(npx)

plt.plot(x, y)
plt.plot(npx, npy)
plt.show()

# magic command (ipython): %timeit
```

```python
# kreslime funkci sin od 0 do 2pi
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
# y2 = []
# for i range(len(y)):
#    val = y[i]
#    if abs(val) > 0.5:
#        y[i] = 0.5 * np.sign(val)

# np.where(condition, true-block, false-block)
y2 = np.where(np.abs(y) > 0.5, 0.5 * np.sign(y), y)
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

```python
while True:
    try:
        a = input("Enter a number greater than 5: ")
        if (a == 'q'):
            break
        if not a.isnumeric():
            raise TypeError("That is not a number!")
        b = int(a)
        if b <= 5:
            raise ValueError("That is not a number greater than 5!")
    except ValueError as ve:
        print(ve)
    except Exception as e:
        print(e)
```
