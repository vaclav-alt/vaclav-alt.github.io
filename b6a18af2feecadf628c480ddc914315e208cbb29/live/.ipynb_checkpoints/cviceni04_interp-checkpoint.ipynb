## Logicke vyrazy

1. implicitni konverze pomoci bool()
2. vysledek logicke operace
 - operatory porovnani: `==`, `!=`, `<`, `>`, `<=`, `>=`
 - logicke operatory: `and`, `or`, `not`
 - membership operatory: `in`, `not in`
 - operatory identity: `is`, `is not`

````python
print(1 == 1)
print(not 1==1)
a = 1
b = 2
print((a == 1) and (b == 1))
print((a == 1) or (b == 1))
````

````python
some_things = ["tuzka", "brokolice", "lenochod", "jogurt", "osciloskop"]

print("brokolice" in some_things)
print("klobasa" not in some_things)
print(not "klobasa" in some_things)
````

````python
a = 2
b = 2
print(id(a) == id(b))
print(a is b)
print(a is not b)
print(not a is b)
````

````python
condition = 1==0
if condition: # implicitni konverze na bool
    print("ano")
else:
    print("ne")
````

````python
# evaluate password quality
password = "lekoricejeobjektivnehnusna"

delka = len(password) # length

if delka < 5:
    print("heslo je moc kratke")
elif delka > 15: # else if, (elsif)
    print("heslo je zbytecne dlouhe")
else:
    print("heslo je ok")
````

## Switch
V mnoha jazycích existuje konstrukce `switch`, která nahrazuje sérii `if-elif-...-elif-else`. Příklad v C:

```c

if (number == 1) {
    do_thing_1();
}
else if (number == 2) {
    do_thing_2();
}
//...
else if (number == k) {
    do_thing_k();
}
else {
    do_default_thing();
}


switch (number) {
    case 1:
        do_thing_1();
        break;
    case 2:
        do_thing_2();
        break;
    // ...
    case k:
        do_thing_k();
        break;
    default:
        do_default_thing();
}
```
V Pythonu to ale neni.

````python
from math import sqrt, exp, log, log2

switcher = {
    1: sqrt,
    2: exp,
    3: log
}

x = 2.0
func_number = 5
func = switcher.get(func_number, log2)
print(func)
print(func(x))
````

````python
user = {
    "name" : "Karl",
    "surname" : "von Bahnhof"
}
print(user.get("phone", 0))
````
