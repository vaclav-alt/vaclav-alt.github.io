## Mutable vs. immutable

- vše je objekt
- objekty seskupují "data" a funkce na nich ("metody")
- každý objekt má `id`, `type`, `value`

```python
a = 1500
print(id(a), type(a), a)
a = a + 1
print(id(a), type(a), a)
```

Immutable typy
- `int`
- `float`
- `bool`
- `string`
- `complex`
- `frozen set`
- `tuple`
- `range`

Mutable typy
- `list`
- `set`
- `dict`

## List (seznam)

```python
colors = ["blau", "rot", "gelb", "schwarz"]
print(type(colors))
print(colors[2])

only_some_colors = colors[1:3]
print(only_some_colors)

colors.append("grau")
colors.append("grau")
print(colors)

colors.pop(4)
print(colors)
```

```python
numbers = [1, 2, 3]
print(id(numbers))
numbers.append(4)
print(id(numbers))

numbers[0] = 0
print(id(numbers))
```

```python
a = 1000
b = a
print(id(a))
print(id(b))
b += 1
print(a, b)
print(id(a), id(b))
```

```python
a = [1, 2, 3]
b = a
print(id(a), id(b))
b.append(4)
print(a, b)
print(id(a), id(b))
```

```python
from numpy import sqrt
sqrt(2)
```

```python
from copy import copy
a = [1, 2, 3]
b = copy(a)
print(id(a), id(b))
b.append(4)
print(a, b)
print(id(a), id(b))
```

```python
from copy import copy, deepcopy
a = [[1,2],[3,4]]
b = deepcopy(a)
b[0][0] = 0
print(a)
```

## Tuple (n-tice)

```python
a = (1, 2)
print(a[0])

print(id(a))
a = (2, 2)
print(id(a))
```

```python
a = [1, 2]
b = tuple(a)
print(b)
c = (a)
print(c)
```

```python
coods = (1.0, -5.0)
```

## Set (množina)

```python
fruit = {"jablko", "hruska", "jahoda", "pomeranc", "papaja"}
other_fruit = {"hruska", "tresen"}
print(fruit)
print("hruska" in fruit)
print(fruit - other_fruit)
print(fruit & other_fruit)
print(fruit | other_fruit)
print(fruit ^ other_fruit)
```

```python
a = set("aaaaahhhoooojjj")
print(a)
```

## Range


```python
a = range(1, 100)

print(a)
b = list(a)
print(b)
```

## Complex

```python
a = 2.0 + 1.0j
print(type(a))
print(a.real, a.imag)
print(abs(a))
b = 1.0j
print(b**2)
```

## Iterace

```python
a = range(10)
b = list(a)
colors = ["blau", "rot", "gelb", "schwarz"]
c = (1, 2)
for i in c:
    print(i)
```

```python
user = {
    # key : value
    "name" : "Erwin Schrodinger",
    "cellphone" : 66612456
}

for x in user:
    print(user[x])
    
for key, value in user.items():
    print(key, value)
```
