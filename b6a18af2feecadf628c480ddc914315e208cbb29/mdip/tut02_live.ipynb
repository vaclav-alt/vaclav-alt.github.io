````python
run_opts = {
    "check_updates" : True,
    "logfile" : "path/to/logfile"
}

def run(check_updates = False, logfile = "log", **kwargs):
    function = kwargs.get("function")
    
    print(f"logging into {logfile}\n")
    
    if check_updates:
        print("checking updates\n")
        
    if function is not None:
        print("running function " + function.__name__)
        function()
    
def func():
    print("neco")
    
run(**run_opts, function = func)
````

## Kontext

````python
x = 1 # globalni

def do():
    x = 1 # lokalni
    x += 1
    print(x)
    
do()
print(x)
````

````python
x = 1

def do():
    global x
    x += 1
    print(x)
    
do()
print(x)
````

````python
k = 1

def do():
    l = 1
    
    def change_k():
        global k
        k += 1
        
    def change_l():
        nonlocal l
        l += 1
        
    print(k, l)
    change_k()
    print(k, l)
    change_l()
    print(k, l)
    
do()
````

````python
def do(f, x): # zjednodusene map
    return f(x)

def program():
    a = 1
    b = 2
    x = 3
    
    def fun(x, a = a):
        # nonlocal a
        a = x + a + b
        return a * x + b

    print(do(fun, x))
    
program()
````

## Nahodna cisla

````python
# linear congruential generator
# x_n = (a * x_(n-1) + c) % m

a = 5
c = 0
m = 256
x0 = 3 # seed
x = x0

def randint():
    global x
    x = (a * x + c) % m
    return x

lst = [randint() for i in range(20)]
print(lst)
````

````python
import random as r

r.seed(3)
lst = [r.randint(0, 10) for i in range(20)]
print(lst)
````

````python
import random as r

state = r.getstate()
lst = [r.randint(0, 10) for i in range(20)]

r.setstate(state)
lst2 = [r.randint(0, 10) for i in range(20)]

print(lst)
print(lst2)
````

````python
# Polynom

def polynom(x, *coefs):
    val = coefs[-1]
    for c in coefs[-2::-1]:
        val *= x
        val += c
    return val

x  = 1 
polynom(x, 3, 2, 1)
````

````python
import numpy as np
import matplotlib.pyplot as plt

args = (3, 0, 1)

def parabola(x):
    return polynom(x, *args)

x = np.linspace(-10, 10, 100)
y = parabola(x)
plt.plot(x, y)
plt.show()
````

````python
def faktorial1(n):
    fak = 1
    for i in range(1, n+1):
        fak *= i
    return fak

def faktorial2(n):
    if n <= 1:
        return 1
    else:
        return n * faktorial2(n-1)
    
faktorial1(0)
print(faktorial2(0))
````
