# Valid string

Napište funkci `is_valid(instr, forbidden)` vracející `True/False`, která rozhodne, zda je řetězec `instr` validní, tj. že neobsahuje znaky z řetězce `forbidden`.

## Řešení

```python
def is_valid(instr, forbidden="!@#$%^&*()"):
    for f in forbidden:
        if f in instr:
            return False
    return True
```

```python
is_valid("abc", forbidden="c")
```
