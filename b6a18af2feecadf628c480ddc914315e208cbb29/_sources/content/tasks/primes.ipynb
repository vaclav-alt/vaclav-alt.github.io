# Prvočísla

Napište funkci `is_prime(num)`, která na vstupu dostane celé číslo a vrátí `True/False`, podle toho, zda je dané číslo prvočíslem. Dále napište funkci `get_primes(numbers)`, která dostane kolekci čísel a vybere z ní prvočísla a navrátí je jako seznam.

## Řešení

```python
def is_prime(n):
    if (n > 1):
        for i in range(2, n):
            if (n % i) == 0:
                return False
        return True
    else:
        return False
    
def get_primes(numbers):
    primes = []
    for i in range(len(numbers)):
        num = numbers[i]
        if (is_prime(num) == True):
            primes.append(num)
        else:
            continue
    return primes
        
```

```python
numbers = [1, 2, 3, 4, 5, 6]

primes = get_primes(numbers)
print(primes)
```

## Trochu čistší řešení

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if (n % i) == 0:
            return False
    return True

def get_primes(numbers):
    primes = []
    for num in numbers:
        if is_prime(num):
            primes.append(num)
    return primes
```

## Very Pythonic řešení

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if (n % i) == 0:
            return False
    return True

def get_primes(numbers):
    return [x for x in numbers if is_prime(x)]
```

## Funkcionální řešení

```python

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if (n % i) == 0:
            return False
    return True

def get_primes(numbers):
    return list(filter(is_prime, numbers))
```
