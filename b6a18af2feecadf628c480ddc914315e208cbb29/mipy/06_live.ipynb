## Nahodna cisla

```python
# linear congruential generator

# x_n = (a*x_n-1 + c) % m

a = 1
c = 0
m = 16
x0 = 3
x = x0 # seed

def randint():
    global a, c, m, x
    x = (a * x + c) % m
    return x

lst = [randint() for i in range(20)]
print(lst)
```

## Sorting

```python
lst = list(range(10, 0, -1))
print(lst)

def compare(a, b):
    return a < b

def swap(lst, i, j):
    lst[i], lst[j] = lst[j], lst[i]
    
def sort(lst, comp):
    # bubble sort (YT: sorting algorithm visualisation)
    n = len(lst)
    for i in range(n-1):
        for j in range(n - i - 1):
            if not comp(lst[j], lst[j+1]):
                swap(lst, j, j+1)

sort(lst, compare)
print(lst)
```

```python
lst = ["a", "noha", "ab", "letadlo", "olihen"]

def strcomp(a, b):
    return len(a) > len(b)
sort(lst, strcomp)
print(lst)
```

```python
prices = {
    "klobasa" : 40,
    "jogurt" : 20,
    "rohlik" : 3
}

nakup = ["jogurt", "klobasa", "rohlik"]

sort(nakup, lambda a,b: prices[a] < prices[b])
print(nakup)
```
