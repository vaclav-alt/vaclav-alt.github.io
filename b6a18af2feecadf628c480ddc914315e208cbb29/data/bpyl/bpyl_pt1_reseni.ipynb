## 1

`lst` je seznam celých čísel (kladných a záporných). Z čísel v listu `lst` vytvořte list `results` tak, že pokud je číslo nezáporné, zapíšete jeho druhou mocninu, a pokud je záporné, nahradíte ho nulou. Učiňte tak v jednom řádku.

Příklad
```
lst = [3, -5, 0, -3, 2]
result = # zde bude vas kod

# results bude obsahovat [9, 0, 0, 0, 4]
```

````python
# reseni
lst = [3, -5, 0, -3, 2]
result = [x**2 if x >= 0 else 0 for x in lst]

print(result)
````

> avg(1, 2, 3)

## 4

Napište funkce `count_vowels(text)`, která spočítá samohlásky (a, e, i, y, o, u) ve stringu `text`.

```
> count_vowels("hello")
2
> count_vowels("bflmpsvz")
0
> count_vowels("aeiyou")
6
```

````python
# reseni

def count_vowels(text):
    count = 0
    for c in text:
        if c in "aeiyou":
            count += 1
    return count

count_vowels("hello")
````

````python
# vystredni reseni (potencialne pametove narocne)

def count_vowels(text):
    return sum([1 if c in "aeiyou" else 0 for c in text])

count_vowels("hello")
````

## 5

V proměnné `raw` je dán řetězec formátu `"key1=val1;key2=val2;key3=val3"`. Převeďte ho do slovníku a výsledek uložte do proměnné `result`.

```
raw = "jmeno=ferdinand;prijmeni=peroutka"

#
# vas postup
#

print(result)

vystup: { "jmeno": "ferdinand", "prijmeni": "peroutka"}
```

````python
# reseni

raw = "jmeno=ferdinand;prijmeni=peroutka"

entries = raw.split(";")

result = {}
for entry in entries:
    key, val = entry.split("=")
    result[key] = val
    
print(result)
````

````python
# alternativni reseni :D

raw = "jmeno=ferdinand;prijmeni=peroutka"

result = { key:val for key, val in [x.split("=") for x in raw.split(";")]}
    
print(result)
````

## 7

Napište funkci `return_only_integers(Sequence)`, která ze zadané sekvence navrátí pouze celá čísla, opět v sekvenci.

```
> return_only_integers([True, "audio jack", 4, 4.3, 1])
[4, 1]
> return_only_integers([2])
[2]
> return_only_integers([1.2, False, "True"])
[]
```

````python
# reseni

def return_only_integers(lst):
    return [x for x in lst if type(x) == int]

return_only_integers([True, "audio jack", 4, 4.3, 1])
````

## 8

Pomocí jednoho řádku prohoďte klíče a hodnoty ve slovníku `x` a uložte do proměnné `result`.

```
x = {"key1" : "val1", "key2" : "val2"}
result = # vas kod

# v result bude { "val1": "key1", "val2": "key2 }
```

````python
# reseni

x = {"key1" : "val1", "key2" : "val2"}

result = {val:key for key, val in x.items()}
print(result)
````

## 9

Odstraňte z řetězce v proměnné `text` všechny výskyty nežádoucích symbolů uvedených v `to_remove`. Výsledke uložte do proměnné `result`.

```
text = "The quick brown fox jumps over a lazy dog."
to_remove = "a!?,.()"

result = # ...

# v result bude: the quick brown fox jumps over  lzy dog
```

````python
# reseni

text = "the quick brown fox jumps over a lazy dog"
to_remove = "a!?,."

result = "".join([c for c in text if c not in to_remove])

print(result)
````

## 10

Napiště funkci `convert`, která přijímá klíčová slova a vrátí list tuplů tvaru `(key, val)`

```
> convert(key1 = "val1", key2 = "val2")
[('key1', 'val1'), ('key2', 'val2')]
```

````python
# reseni

def convert(**kwargs):
    return [(key, val) for key, val in kwargs.items()]

convert(key1 = "val1", key2 = "val2")
````
