## Faktorial

$$
n! = n (n-1) (n-2) \cdots 2\cdot 1
$$

$$
0! = 1
$$

```python
def faktorial(n):
    if n == 0:
        return 1
    else:
        return n * faktorial(n-1)
    
faktorial(5)
```

## Dictionary a pretty print

```python
d = dict(
    jmeno = "vaclav",
    prijmeni = "alt",
    adresa = dict(
        ulice = "na konci",
        cislo = "1"
    )
)

def pretty_print(d, w = " ", indent = 0):
    white = w * indent
    for key, val in d.items():
        if type(val) == dict:
            print(white + f"{key}:")
            pretty_print(val, w, indent + 1)
        else:
            print(white + f"{key}: {val}")
    
pretty_print(d, "\t", 2)
    
# jmeno: vaclav
# prijmeni: alt
# adresa:
#     ulice: na konci
#     cislo: 1
```

## Bisekce

```python
a = 1
b = 8

def f(x):
    return x**2 - 2

eps = 1e-15

i = 0
while abs(a-b) > eps:
    i += 1
    c = a + (b-a) * 0.5
    
    if f(a) * f(c) < 0:
        b = c
    elif f(c) * f(b) < 0:
        a = c
    else:
        print("bad boundaries")
        exit()
        
a, i
```

```python
def f(x):
    return x**2 - 2

def bisection(f, a, b, eps = 1e-8):
    i = 0
    while abs(a-b) > eps:
        i += 1
        c = a + (b-a) * 0.5

        if f(a) * f(c) < 0:
            b = c
        elif f(c) * f(b) < 0:
            a = c
        else:
            print("bad boundaries")
            return
    
    return a, i

result = bisection(f, 6, 8)
result
```
