## Prvocisla

Napiste funkci is_prime(x), ktera overi, ze zadane cislo x je prvocislem. Navraci hodnoty True/False

```python
# operace modulo % - zbytek po celociselnem deleni

6 % 2
```

```python
from math import sqrt

def is_prime(x):
    if x <= 1:
        return False
    for i in range(2, int(sqrt(x))+1):
        if (x % i) == 0:
            return False
        
    return True

is_prime(4)
```

```python
x = int(2.4)
type(x), x
```

```python
numbers = [1,2,3,4,5,6]

primes = []

for i in range(len(numbers)):
    num = numbers[i]
    if is_prime(num):
        primes.append(num)
        
print(primes)
```

```python
numbers = [1,2,3,4,5,6]

primes = []

for num in numbers:
    if is_prime(num):
        primes.append(num)
        
print(primes)
```

```python
# list comprehensions

numbers = [1,2,3,4,5,6]

primes = [ num for num in numbers if is_prime(num)]
        
print(primes)
```

```python
numbers = [1,2,3,4,5,6]

def get_primes(nums):
    for num in nums:
        if not is_prime(num):
            nums.remove(num)
    return nums

primes = get_primes(numbers[:])

print(numbers)
print(primes)
```

## Mutable vs immutable

immutable: int, float, bool, string, complex, tuple, range
mutable: list, set, dict

```python
x = 1
y = 1
x is y # to je to same jako id(x) == id(y)
```

```python
x = 1
print(id(x))
x += 1
print(id(x))
```

```python
lst = [1, 2]
print(id(lst))
lst.append(3)
print(id(lst))
print(id(lst[0]))
lst[0] = 2
print(id(lst[0]))
```

```python
lst1 = [1, 2]
lst2 = lst1[:] # shallow copy
lst1 is lst2
```

```python
lst = [
    [1,2],
    [3,4]
]
lst2 = lst[:]
lst2[0][0] = 2
print(lst, lst2)
```

```python
from copy import deepcopy

lst = [
    [1,2],
    [3,4]
]
lst2 = deepcopy(lst)
lst2[0][0] = 2
print(lst, lst2)
```

## Řetězce

Implementujte funkci is_valid(instr, forbidden), která ověří, že v řetězci instr se nenachází žádný ze znaků v řetězci forbidden. Funkce vrací True/False.

```python
forbidden = "!@#$%^&*()"

def is_valid(instr, forbidden):
    for x in instr:
        for f in forbidden:
            if x == f:
                return False
            
    return True

is_valid("", forbidden)        
```

```python
forbidden = "!@#$%^&*()"

def is_valid(instr, forbidden):
    for x in instr:
        if x in forbidden:
            return False
            
    return True

is_valid("abc", forbidden)
```

```python
def is_numeric(instr):
    for x in instr:
        if x not in "1234567890":
            return False
    return True

is_numeric("123"), "123".isdigit()
```
